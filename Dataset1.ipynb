```python
import pandas as pd

train = pd.read_csv("train.csv")
val = pd.read_csv("val.csv")

train.head()
val.head()
```

```python
print("Missing values in train:")
print(train.isnull().sum())

print("\nMissing values in val:")
print(val.isnull().sum())
```

```python
print("Train category distribution:")
print(train['category'].value_counts())

print("\nValidation category distribution:")
print(val['category'].value_counts())
```

```python
import matplotlib.pyplot as plt

train['category'].value_counts().plot(kind='bar', title='Train Dataset - Category Distribution')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

```python
val['category'].value_counts().plot(kind='bar', title='Validation Dataset - Category Distribution')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

sample = train.sample(5, random_state=42)

fig, axes = plt.subplots(1, 5, figsize=(15,5))
for i, row in enumerate(sample.itertuples()):
    img_path = base_path + row._1
    img = mpimg.imread(img_path)
    axes[i].imshow(img)
    axes[i].set_title(f"Category: {row.category}")
    axes[i].axis('off')
plt.show()
```
